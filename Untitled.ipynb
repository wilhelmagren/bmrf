```python
assets = [
    'Revolution Case', 'Dreams & Nightmares Case', 'Clutch Case',
    'Fracture Case', 'Glove Case', 'Recoil Case', 'Prisma 2 Case',
    'Prisma Case', 'Operation Bravo Case', 'CS:GO Weapon Case', 'Snakebite Case',
    'Operation Breakout Weapon Case', 'Operation Hydra Case', 'Danger Zone Case',
    'Huntsman Weapon Case', 'Winter Offensive Weapon Case', 
    'Operation Vanguard Weapon Case', 'CS:GO Weapon Case 3', 'Chroma Case',
    'Chroma 2 Case', 'Chroma 3 Case', 'Horizon Case', 'Operation Wildfire Case',
    'Spectrum Case', 'Gamma 2 Case',
]

n_assets = len(assets)
print(n_assets)
```

```python
data = {}

for asset in assets:
    url = 'http://steamcommunity.com/market/pricehistory'
    history = requests.get(
        url, 
        params={
        'appid': appid,
        'market_hash_name': asset,
        'country': country,
        'currency': currency,
      },
        cookies={
            'steamLoginSecure': steamLoginSecure,
        }
    )
    data[asset] = history.json()['prices']
```

```python
import numpy as np
dataset = {}
for asset in assets:
    dataset[asset] = {}
    dataset[asset]['price'] = np.array([p[1] for p in data[asset]])
    dataset[asset]['time'] = np.array([p[0].split(':')[0] for p in data[asset]])
    
```

```python
from datetime import datetime
import matplotlib.dates
import calendar

m2n = {m: i for i, m in enumerate(calendar.month_abbr) if month}
del[m2n['']]

for asset in assets:
    dates = []
    for t in dataset[asset]['time']:
        month, day, year, time = t.split(' ')
        dtime = datetime(int(year), m2n[month], int(day), int(time))
        dates.append(dtime)
    plt.plot(matplotlib.dates.date2num(dates), dataset[asset]['price'], label=asset)

plt.title('Assets price over time')
plt.ylabel('Price [€]')
plt.xlabel('Time')
#plt.legend(loc='best')
plt.show()
```

```python
from qiskit_finance.applications.optimization import PortfolioOptimization

means = []
for asset in assets:
    means.append(dataset[asset]['price'].mean())

means = np.array(means)
print(means)
```

```python
time_lens = []
for asset in assets:
    l = len(dataset[asset]['time'])
    time_lens.append((asset, l))
    
time_assets = list(reversed(sorted(time_lens, key=lambda x: x[1])))
```

```python
padded_dataset = {}
required_len = time_assets[0][1]
longest_asset = time_assets[0][0]

for asset in assets:
    ll = len(dataset[asset]['price'])
    padded_dataset[asset] = {}
    padded_dataset[asset]['price'] = np.concatenate(([0.03 for _ in range(required_len - ll)], dataset[asset]['price']))
    padded_dataset[asset]['time'] = dataset[longest_asset]['time']
    
```

```python
for asset in assets:
    assert len(padded_dataset[asset]['price']) == required_len
    
```

```python
prices = []
for asset in assets:
    prices.append(padded_dataset[asset]['price'])
    
prices = np.concatenate([prices], axis=0)
```
